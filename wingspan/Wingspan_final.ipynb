**Distribution of Students Across Years:** The histogram shows how students are distributed across different academic years. The height of each bar corresponds to the number of students in each year.

**Overall Trend:** The shape of the histogram gives a sense of whether there's a balance in the number of students across years or if certain years dominate. For instance, a right-skewed histogram would suggest a greater number of lower-year students.

```python
# Example: Box plot for a numerical variable
data.boxplot(column=['Year of Study'])
plt.title('Box Plot of Year of Study')
plt.show()
```

**Median Year of Study:** The line within the box shows the median year of study. This is the year that divides the dataset into two equal halves.

**Interquartile Range (IQR):** The length of the box represents the IQR, indicating the spread of the middle 50% of the data. A larger box would suggest a greater diversity in the years of study among respondents.

**Outliers:** If there are any dots (individual points) outside the whiskers, these represent years that are unusually high or low in terms of the number of students compared to the rest of the data.

**Data Spread and Symmetry:** The box plot provides a visual representation of the spread of the data. If the box is symmetric, it suggests an even distribution around the median. If the box is asymmetric, it indicates skewness in the year distribution.

**Extremes in Data:** The ends of the whiskers represent the range of the data, excluding outliers. This gives a sense of the overall range of the 'Year of Study' in your dataset.

```python
import difflib
import nltk
from nltk.corpus import stopwords

# Download the set of stop words the first time
nltk.download('stopwords')

# Then you can load the stop words for a specific language, for example, English
stop_words = set(stopwords.words('english'))


root_words = [
    'PUNUGULU', 'MIRCHI', 'ALOO BAJJI', 'PAKODI', 'CORN SAMOSA', 'NUGGETS', 'FRENCH FRIES',
    'MASALA VADA', 'DAL VADA', 'PAROTA CHICKEN', 'CHICKEN BIRYANI', 'EGG BIRIYANI', 'EGG PAROTA',
    'EGG CHAPATHI', 'MEALS', 'CHOLE POORI', 'CHAPATHI VEG', 'PAROTA VEG', 'GOBI RICE', 'VEG PULAV',
    'VEG FRIED RICE', 'TOMATO RICE', 'JEERA RICE', 'VEG MANCHURIA', 'VADA', 'BONDA', 'POORI',
    'MASALA DOSA', 'ONION DOSA', 'UTHAPPAM', 'PLAIN DOSA', 'UPMA', 'COFFEE', 'LEMON TEA', 'HORLICKS',
    'BOOST', 'BADAM MILK', 'ROSE MILK', 'MASALA TEA', 'HOT CHOCO', 'TEA', 'IDLY', 'CHAPATHI CHICKEN'
]



# Function to find the closest root word using difflib's sequence matcher
def closest_root_word_difflib(response, root_words):
    response = response.lower().strip()
    # Using difflib to find the closest match
    closest_match = difflib.get_close_matches(response, [word.lower() for word in root_words], n=1, cutoff=0.7)

    # Return the match if found, else return "No Match"
    return closest_match[0].upper() if closest_match else "No Match"

# Applying the improved mapping function to each relevant column
for column in free_data.columns[:]:
    free_data[column + "_Mapped_Improved"] = free_data[column].astype(str).apply(lambda x: closest_root_word_difflib(x, root_words))

# Display the modified dataset with improved mappings
free_data.head()
```

# Model Building

```python
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report

# Assuming 'cluster' is your target variable and all other columns are features
X = result.drop('cluster', axis=1)
y = result['cluster']

# One-hot encoding of categorical variables
one_hot_encoder = OneHotEncoder()
X_encoded = one_hot_encoder.fit_transform(X)

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Training the LightGBM Classifier
lgb_classifier = lgb.LGBMClassifier(random_state=42)
lgb_classifier.fit(X_train, y_train)

# Predicting on the test set
y_pred = lgb_classifier.predict(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
classification_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(classification_report)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier


# Hyperparameter tuning with GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                           param_grid=param_grid,
                           cv=3,
                           n_jobs=-1,
                           verbose=2,
                           scoring='accuracy')

grid_search.fit(X_train, y_train)

# Best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

best_params, best_score
```
